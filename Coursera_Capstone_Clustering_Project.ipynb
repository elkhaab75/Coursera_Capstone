# PART I: 

## 1. Download Data from Wikipedia Page


To download the Toronto Neiborhoods dataset from the Wiki page I opted to use pandas library for simplicity. I also tried using the Beautiful Soup package but that would have required a lot of coding efforts to pull the neighborhoods table.

1. Read html using the url provided.
2. Check how many tables are in the Wiki page
3. By visiting the page the first table is the neighborhoods table, hence, that our pick to construct our toronto neighborhoods dataframe

```python
# Scrapping the Toronto Neigborhoods Wiki Page and assigning the first table to our dataframe
wiki_url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

tables = pd.read_html(wiki_url)
print("There are : ",len(tables)," tables")
print("Take look at table 0")
print(tables[0])
toronto_data = tables[0]
```

## 3. Pre-Processing our Dataset

1. Ignore Cells with Borough is <b> Not assigned </b>
2. Make sure that Postal Code is not listed twice; hence, group neighborhoods into their corresponding Postal Code and seperate them with a comma.
3. If a cell has a borough but <b> Not assigned </b> then the neighborhood will be the same as the borough.
4. Display the new number of rows in the dataframe.

```python
# Removing Not Assigned Borough from our dataset
toronto_data = toronto_data[toronto_data['Borough'] != 'Not assigned'].reset_index(drop=True)
toronto_data.head(15)
```

```python
# Number of rows after exclusion of Not assigned Borough
toronto_data.shape
```

```python
# Make sure that neighborhoods are grouped by Postal Code
toronto_data.groupby(by=['Postal Code'])
```

```python
toronto_data.head(15)
```

```python
# Replace any Not assigned neigborhood Borough value.
toronto_data['Neighbourhood'] = toronto_data['Neighbourhood'].replace('Not assigned', toronto_data['Borough'])
```

# PART II:

## Assign Latitude and Longtitude

Now that we build a dataframe of postal code of each neighborhood along with the borough name and neigborhood name, we will need to assign latitude and longitude coordinates for each neighborhood. 
To do that, I originally tried using geocoder, however, I couldn't get any data after several tries; hence, I went with reading the latitude and longitude from the csv file provided. Once, successful, I merged both dataframes based on Postal Code so I can enrish toronto_data dataframe to obtain the final dataframe with latitude and longitude.


```python
# read the geographical coordinates from the csv into a dataframe
geo_coordinates= pd.read_csv('http://cocl.us/Geospatial_data')
geo_coordinates.head()
```

```python
# Merge both dataframes to obtain the final toronto data neighborhoods dataframe with the correct latitude and longitude
toronto_data_final = toronto_data.merge(geo_coordinates, on="Postal Code", how = 'inner')
toronto_data_final.head(20)
```

```python
# Get the number of rows in the final toronto neighborhood dataframe
toronto_data_final.shape
```

```python
import requests

# Function to get near by venues- This will be called to get near by venues in Toronto Neighborhoods.
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
#Now write the code to run the above function on each neighborhood and create a new dataframe called toronto_venues.

# type your answer here
toronto_venues = getNearbyVenues(names=toronto_df['Neighbourhood'],
                                   latitudes=toronto_df['Latitude'],
                                   longitudes=toronto_df['Longitude']
                                  )


print(toronto_venues.shape)
```
